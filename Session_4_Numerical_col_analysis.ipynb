**emperical rule**

```python
pwage=visa_df['prevailing_wage']
Mean=np.mean(pwage)
Std=np.std(pwage)
lower_bound = Mean-(1*Std)
upper_bound = Mean+(1*Std)
```

```python
lower_bound,upper_bound
con1=pwage>lower_bound
con2=pwage<upper_bound
```

```python
con=con1&con2
```

```python
len(visa_df[con])==68*25480/100
```

```python
pwage=visa_df['prevailing_wage']
Mean=np.mean(pwage)
Std=np.std(pwage)
lower_bound = Mean-(2*Std)
upper_bound = Mean+(2*Std)
lower_bound,upper_bound
con1=pwage>lower_bound
con2=pwage<upper_bound
con=con1&con2
len(visa_df[con])==95*25480/100
```

```python
pwage=visa_df['prevailing_wage']
Mean=np.mean(pwage)
Std=np.std(pwage)
lower_bound = Mean-(3*Std)
upper_bound = Mean-(3*Std)
lower_bound,upper_bound
con1=pwage>lower_bound
con2=pwage<upper_bound
con=con1&con2
len(visa_df[con])==99.7*25480/100
```
