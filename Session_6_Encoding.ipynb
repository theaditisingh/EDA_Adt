**Encoding**

- Encoding means convert categorical data to Numerical data

- ML/DL or any model works on maths

- Models can not understand english characters

- So it is very important convert categorical label to numerical values

- For example Gender column has two labels

    - Male
 
    - Female
 
- Then Female as O, Male represent as 1 based on alphabetical order

```python
from sklearn.preprocessing import LabelEncoder
```

- Step-1: import the method

- Step-2: save the method

- Step-3: apply fit transform

    - **fit**: means find the logic

        - In above example fit means create **mapper** or **dict**
 
    - **transform**: apply the logic and change the data
 
        - apply the **mapper on the data**

```python
data=r"C:\Users\singh\Downloads\Visadataset.csv"
visa_df=pd.read_csv(data)
cat=visa_df.select_dtypes(include='object').columns

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
visa_df['case_status']=le.fit_transform(visa_df['case_status'])
visa_df
```
