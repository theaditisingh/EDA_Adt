```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=r"C:\Users\singh\Downloads\Visadataset.csv"
visa_df=pd.read_csv(data)

cat=visa_df.select_dtypes(include='object').columns
num=visa_df.select_dtypes(exclude='object').columns
```

```python
pwage = visa_df['prevailing_wage']
Mean = np.mean(pwage)
Std = np.std(pwage)
Nr = pwage - Mean
z_Wage = Nr/Std
visa_df['z_Wage'] = z_Wage
visa_df
```

```python
wage_data = visa_df['prevailing_wage']
Mean = np.mean(wage_data)
Std = np.std(wage_data)
Nr = wage_data - Mean
z_Wage = Nr/Std
visa_df['z_Wage'] = z_Wage
```

```python
visa_df[['prevailing_wage','z_Wage']]
```

```python
plt.figure(figsize=(10,6))
plt.suptitle('Raw-data vs Z-data')
plt.subplot(2,2,1).hist(wage_data)
plt.title('prevailing_wage')
plt.subplot(2,2,2).hist(z_Wage)
plt.title('Z score of prevailing_wage')
plt.subplot(2,2,3).boxplot(wage_data,vert=False)
plt.subplot(2,2,4).boxplot(z_Wage,vert=False)
plt.show()
```

**MinMaxScaler**

```python
# Step-1:read the data
# Step-2: calculate min
# Step-3: calculate maxm
# Step-4: Nr=step1-step2
# Step-5: Dr=Step3-Step2
# Step-6: Nr/Dr
# Step-7: visa_df['wage_min_max']=Nr/Dr
```

```python
wage_data = visa_df['prevailing_wage']
wage_min=wage_data.min()
wage_max=wage_data.max()
Nr=wage_data-wage_min
Dr=wage_max-wage_min
Nr/Dr
visa_df['wage_min_max']=Nr/Dr
```
